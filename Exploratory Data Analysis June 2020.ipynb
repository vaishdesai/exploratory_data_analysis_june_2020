SCORE ANALYSIS:

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.distplot(stu_data['math score'],kde=False,color='b',bins=30)
plt.ylabel('Frequency')
plt.title('Math scores of Students')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.distplot(stu_data['reading score'],kde=False,color='b',bins=30)
plt.ylabel('Frequency')
plt.title('Reading scores of Students')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.distplot(stu_data['writing score'],kde=False,color='b',bins=30)
plt.ylabel('Frequency')
plt.title('Writing scores of Students')
```

We will study the marks in these subjects against all parameters given in the data

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="gender",y="math score",data=stu_data)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="gender",y="reading score",data=stu_data)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="gender",y="writing score",data=stu_data)
plt.show()


```

FRom these bar graphs it is clear that male population performed well in the math test whereas the female population did well in the reading and writing test.

Now we will try to understand how parental level of education has affected the scores of students in all the three tests.

```python
from matplotlib.pyplot import xticks
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.stripplot(x="parental level of education",y="math score",data=stu_data)
plt.xticks(rotation=90)
plt.subplot(1,3,2)
sns.stripplot(x="parental level of education",y="reading score",data=stu_data)
plt.xticks(rotation=90)
plt.subplot(1,3,3)
sns.stripplot(x="parental level of education",y="writing score",data=stu_data)
plt.xticks(rotation=90)

```

From the above graphs, we can observe a common feature that the lowest of scores in each subject are from students whose parents have attended some high school(or from some college)

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
day_wise=stu_data.groupby('race/ethnicity')['math score'].mean().plot.bar()
plt.title("Math scores Vs Race/Ethnicity")
plt.ylabel('Average Math scores')
plt.xlabel('Race/Ethnicity')
plt.subplot(1,3,2)
day_wise=stu_data.groupby('race/ethnicity')['reading score'].mean().plot.bar()
plt.title("Reading scores Vs Race/Ethnicity")
plt.ylabel('Average Reading scores')
plt.xlabel('Race/Ethnicity')
plt.subplot(1,3,3)
day_wise=stu_data.groupby('race/ethnicity')['writing score'].mean().plot.bar()
plt.title("Writing scores Vs Race/Ethnicity")
plt.ylabel('Average Writing scores')
plt.xlabel('Race/Ethnicity')
```

From the above output we can see that Group E has performed comparatively better than all the other groups

Now we will compare same data but with an added factor of if the student has taken a test preparation course or not.

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="race/ethnicity",y="math score",hue="test preparation course",data=stu_data)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="race/ethnicity",y="reading score",hue="test preparation course",data=stu_data)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="race/ethnicity",y="writing score",hue="test preparation course",data=stu_data)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="parental level of education",y="math score",hue="test preparation course",data=stu_data)
plt.xticks(rotation=90)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="parental level of education",y="reading score",hue="test preparation course",data=stu_data)
plt.xticks(rotation=90)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="parental level of education",y="writing score",hue="test preparation course",data=stu_data)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x=stu_data['test preparation course'],y=stu_data['math score'])
plt.subplot(1,3,2)
sns.boxplot(x=stu_data['test preparation course'],y=stu_data['reading score'])
plt.subplot(1,3,3)
sns.boxplot(x=stu_data['test preparation course'],y=stu_data['writing score'])
```

From the above data we can see that those who have completed the course have performed well across all three tests as compared to those who have not.
